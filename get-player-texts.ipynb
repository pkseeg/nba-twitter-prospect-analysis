```python
import pandas as pd
import datetime
import preprocessor as p
```

```python
tweets = pd.read_csv('data/player_tweets.csv')
# remove all the tweets where the 'time' column makes no sense (i.e. NaN)
#   there are only 10 of these tweets
def is_time(t):
    return len(str(t)) == 19
set([len(str(t)) for t in tweets['time']])
tweets = tweets[tweets['time'].map(is_time)]
tweets['datetime'] = pd.to_datetime(tweets['time'])
def get_season(datetime):
    season = str(datetime.year)
    if datetime.month <= 6:
        season = str(datetime.year-1)+'-'+season
    else:
        season += '-'+str(datetime.year+1)
    return season
tweets['season'] = [get_season(x) for x in tweets['datetime']]
tweets.sample(5)
```

This notebook will use the player tweet data and create a new table of player text (by year)

```python
def aggregate_text(s_subset):
    return p.clean(' '.join(s_subset['text'])).lower()
```

```python
def get_player_text(tweets, handle):
    player_text = pd.DataFrame(columns=['year','twitter_handle','cleaned_text'])
    i = 0
    player_tweets = tweets[tweets['handle']==handle]
    seasons = set(player_tweets['season'])
    for season in seasons:
        s_subset = player_tweets[player_tweets['season']==season]
        row = [season,handle,aggregate_text(s_subset)]
        player_text.loc[i] = row
        i += 1
    return player_text
```

```python
def get_all_texts(tweets):
    text_by_season = pd.DataFrame(columns=['year','twitter_handle','cleaned_text'])
    handles = set(tweets['handle'])
    total = len(handles)
    for i, handle in enumerate(handles):
        player_text = get_player_text(tweets,handle)
        print('Processed '+str(handle) + ' '+str(i+1)+'/'+str(total))
        text_by_season = pd.concat([text_by_season,player_text])
    return text_by_season
```
