```python
import pandas as pd
```

```python
nlu_scores = pd.read_csv('data/nlu_scores_final.csv')
nlu_scores.columns
```

```python
nlu_scores.head(2)
```

```python
# topic dataframe will contain
#  columns for all topics
#  twitter_handle, season, Player
all_topics = {}
for i, doc_label in enumerate(nlu_scores['doc_labels']):
    topic_list = doc_label.split(' /')
    topic_list[0] = topic_list[0][1:]
    for topic in topic_list:
        all_topics[topic] = [0]
print(len(all_topics))
```

```python
def get_topic_scores(all_topics, doc_labels, doc_label_weights):
    player_topics = all_topics.copy()
    topic_list = doc_labels.split(' /')
    scores_list = doc_label_weights.split()
    topic_list[0] = topic_list[0][1:]
    for i in range(len(topic_list)):
        player_topics[topic_list[i]] = [int(scores_list[i])]
    return player_topics
```

```python
topic_scores = pd.DataFrame.from_dict(all_topics)
topic_scores['Player'] = ['empty']
topic_scores['season'] = ['empty']
topic_scores['twitter_handle'] = ['empty']
for i in range(len(nlu_scores)):
    player = nlu_scores['Player'].iloc[i]
    season = nlu_scores['season'].iloc[i]
    twitter_handle = nlu_scores['twitter_handle'].iloc[i]
    doc_labels = nlu_scores['doc_labels'].iloc[i]
    doc_labels_weights = nlu_scores['doc_labels_weights'].iloc[i]
    player_topics = get_topic_scores(all_topics, doc_labels, doc_labels_weights)
    player_topics['Player'] = [player]
    player_topics['season'] = [season]
    player_topics['twitter_handle'] = [twitter_handle]
    subdf = pd.DataFrame.from_dict(player_topics)
    topic_scores = pd.concat([topic_scores, subdf], sort=True)
```
