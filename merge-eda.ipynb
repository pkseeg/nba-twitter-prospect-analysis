## IBM Measurements by Catch-Alls
How do different personality measurements correlate with catch-all metrics like VORP, WS, during the same season (i.e. not pre-draft seasons like the models we’re building)?

```python
from scipy import stats
def significant_measurements(df, measure_cols, target_cols, sig_val):
    corr = []
    for m in measure_cols:
        for t in target_cols:
            r, p = stats.pearsonr(df[m],df[t])
            if p < sig_val:
                corr.append((m,t))
    return corr
```

```python
measure_cols = ['emot_anger','emot_disgust','emot_fear','emot_joy','emot_sadness','sent_score','need_challenge', 'need_closeness', 'need_curiosity',
       'need_excitement', 'need_harmony', 'need_ideal', 'need_liberty',
       'need_love', 'need_practicality', 'need_self_expression',
       'need_stability', 'need_structure', 'facet_adventurousness',
       'facet_artistic_interests', 'facet_emotionality', 'facet_imagination',
       'facet_intellect', 'facet_liberalism', 'facet_achievement_striving',
       'facet_cautiousness', 'facet_dutifulness', 'facet_orderliness',
       'facet_self_discipline', 'facet_self_efficacy', 'facet_activity_level',
       'facet_assertiveness', 'facet_cheerfulness', 'facet_excitement_seeking',
       'facet_friendliness', 'facet_gregariousness', 'facet_altruism',
       'facet_cooperation', 'facet_modesty', 'facet_morality',
       'facet_sympathy', 'facet_trust', 'facet_anger', 'facet_anxiety',
       'facet_depression', 'facet_immoderation', 'facet_self_consciousness',
       'facet_vulnerability', 'value_conservation', 'value_openness_to_change',
       'value_hedonism', 'value_self_enhancement', 'value_self_transcendence']
target_cols = ['Yrs','WS','WS/48','BPM','VORP']
```

```python
corrs = significant_measurements(merged, measure_cols, target_cols, 0.001)
print('Significant correlation for '+str(round(100*(len(corrs) / (len(measure_cols) * len(target_cols))),2))+'% of measurements.')
```

```python
# Which ones were best?
best_measures = {}
for c in corrs:
    if c[0] in best_measures:
        best_measures[c[0]] += 1
    else:
        best_measures[c[0]] = 1
best_measures
```

```python
# Let's look at the measures which were significant in all 5 target cols
def plot_sigs(corrs, df, num_targets = 5):
    best_measures = {}
    for c in corrs:
        if c[0] in best_measures:
            best_measures[c[0]] += 1
        else:
            best_measures[c[0]] = 1
    # plot them
    for c in corrs:
        if best_measures[c[0]] == num_targets:
            r, p = stats.pearsonr(df[c[0]],df[c[1]])
            plt.scatter(df[c[0]], df[c[1]])
            plt.xlabel(c[0])
            plt.ylabel(c[1])
            plt.title(c[0]+' by '+c[1]+', r = '+str(round(r,2))+', p = '+str(round(p,4)))
            plt.show()
```

```python
plot_sigs(corrs, merged, 5)
```
